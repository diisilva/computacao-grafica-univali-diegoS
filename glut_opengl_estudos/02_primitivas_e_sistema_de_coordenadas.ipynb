# Aula 02 — Primitivas e Sistema de Coordenadas (NDC)

> Objetivo: entender como o OpenGL desenha formas básicas no plano 2D.

## O que você aprende aqui
- `GL_POINTS`, `GL_LINES` e `GL_TRIANGLES`.
- Como pensar em NDC: X e Y no intervalo `[-1, 1]`.
- Como configurar tamanho de ponto (`glPointSize`) e espessura de linha (`glLineWidth`).

## Roteiro de execução
1. Execute a célula 1 para carregar funções de desenho.
2. Execute a célula 2 para abrir a janela e renderizar a cena.

```python
from OpenGL.GL import *
from OpenGL.GLUT import *

def configure_close_behavior() -> bool:
    """Tenta evitar encerramento do kernel ao fechar a janela GLUT."""
    try:
        glutSetOption(GLUT_ACTION_ON_WINDOW_CLOSE, GLUT_ACTION_GLUTMAINLOOP_RETURNS)
        print("GLUT configurado para retornar ao fechar a janela.")
        return True
    except Exception:
        print("GLUT sem suporte a retorno no close; fechar pode encerrar o kernel.")
        return False

def draw_axes() -> None:
    """Desenha eixos de referência X e Y."""
    glLineWidth(2.0)
    glBegin(GL_LINES)
    glColor3f(0.7, 0.7, 0.7)
    glVertex2f(-1.0, 0.0); glVertex2f(1.0, 0.0)
    glVertex2f(0.0, -1.0); glVertex2f(0.0, 1.0)
    glEnd()

def draw_primitives() -> None:
    """Mostra pontos, linhas e triângulo colorido."""
    glPointSize(10.0)
    glBegin(GL_POINTS)
    glColor3f(1.0, 0.9, 0.2)
    glVertex2f(-0.7, 0.6); glVertex2f(-0.5, 0.45); glVertex2f(-0.3, 0.7)
    glEnd()

    glLineWidth(4.0)
    glBegin(GL_LINES)
    glColor3f(0.2, 0.9, 1.0)
    glVertex2f(-0.1, 0.6); glVertex2f(0.4, 0.75)
    glVertex2f(-0.1, 0.4); glVertex2f(0.4, 0.2)
    glEnd()

    glBegin(GL_TRIANGLES)
    glColor3f(0.9, 0.3, 0.3); glVertex2f(0.55, 0.2)
    glColor3f(0.3, 0.9, 0.4); glVertex2f(0.9, 0.2)
    glColor3f(0.3, 0.4, 1.0); glVertex2f(0.72, 0.7)
    glEnd()

def display() -> None:
    glClear(GL_COLOR_BUFFER_BIT)
    draw_axes()
    draw_primitives()
    glFlush()
```

```python
glutInit()
close_safe = configure_close_behavior()
glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB)
glutInitWindowSize(900, 600)
glutCreateWindow(b"Aula 02 - Primitivas e NDC")
glClearColor(0.08, 0.08, 0.10, 1.0)
glutDisplayFunc(display)

try:
    glutMainLoop()
except SystemExit:
    print("GLUT chamou SystemExit ao fechar a janela.")

if close_safe:
    print("Janela fechada e kernel deve continuar ativo.")
else:
    print("Se o kernel cair, rode esta aula via script .py.")
```
