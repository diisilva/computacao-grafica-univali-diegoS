# Aula 03 — Transformações 2D e Animação com Timer

> Objetivo: aplicar translação, rotação e escala, e animar com `glutTimerFunc`.

## Conceitos importantes
- **Translação**: move o objeto no espaço da cena.
- **Rotação**: gira o objeto em torno de um eixo.
- **Escala**: altera o tamanho do objeto.
- **Timer**: atualiza estado e pede novo frame periodicamente.

## Roteiro de execução
1. Execute a célula 1 (define estado e callbacks).
2. Execute a célula 2 (inicializa GLUT e inicia animação).

```python
from OpenGL.GL import *
from OpenGL.GLUT import *

angle_degrees = 0.0

def configure_close_behavior() -> bool:
    """Tenta evitar encerramento do kernel ao fechar a janela GLUT."""
    try:
        glutSetOption(GLUT_ACTION_ON_WINDOW_CLOSE, GLUT_ACTION_GLUTMAINLOOP_RETURNS)
        print("GLUT configurado para retornar ao fechar a janela.")
        return True
    except Exception:
        print("GLUT sem suporte a retorno no close; fechar pode encerrar o kernel.")
        return False

def draw_triangle() -> None:
    """Desenha um triângulo centrado na origem local."""
    glBegin(GL_TRIANGLES)
    glColor3f(1.0, 0.2, 0.2); glVertex2f(0.0, 0.3)
    glColor3f(0.2, 1.0, 0.2); glVertex2f(-0.3, -0.25)
    glColor3f(0.2, 0.4, 1.0); glVertex2f(0.3, -0.25)
    glEnd()

def display() -> None:
    glClear(GL_COLOR_BUFFER_BIT)
    glMatrixMode(GL_MODELVIEW)
    glLoadIdentity()

    glPushMatrix()
    glTranslatef(0.45, 0.0, 0.0)
    glRotatef(angle_degrees, 0.0, 0.0, 1.0)
    glScalef(1.2, 1.2, 1.0)
    draw_triangle()
    glPopMatrix()

    glutSwapBuffers()

def update_timer(_: int) -> None:
    global angle_degrees
    angle_degrees = (angle_degrees + 1.5) % 360.0
    glutPostRedisplay()
    glutTimerFunc(16, update_timer, 0)
```

```python
glutInit()
close_safe = configure_close_behavior()
glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB)
glutInitWindowSize(900, 600)
glutCreateWindow(b"Aula 03 - Transformacoes e Timer")
glClearColor(0.03, 0.03, 0.05, 1.0)
glutDisplayFunc(display)
glutTimerFunc(0, update_timer, 0)

try:
    glutMainLoop()
except SystemExit:
    print("GLUT chamou SystemExit ao fechar a janela.")

if close_safe:
    print("Janela fechada e kernel deve continuar ativo.")
else:
    print("Se o kernel cair, rode esta aula via script .py.")
```
