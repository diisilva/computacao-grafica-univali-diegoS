# Aula 05 — Reshape, Viewport e Projeção

> Objetivo: manter proporção visual correta quando a janela é redimensionada.

## Conceitos importantes
- `reshape(width, height)` é chamado ao redimensionar a janela.
- `glViewport` define a área de renderização em pixels.
- `gluOrtho2D` ajusta a projeção ortográfica sem distorção.

## Roteiro de execução
1. Execute a célula 1 para definir funções.
2. Execute a célula 2 para abrir a janela e testar resize manualmente.

```python
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *

def configure_close_behavior() -> bool:
    """Tenta evitar encerramento do kernel ao fechar a janela GLUT."""
    try:
        glutSetOption(GLUT_ACTION_ON_WINDOW_CLOSE, GLUT_ACTION_GLUTMAINLOOP_RETURNS)
        print("GLUT configurado para retornar ao fechar a janela.")
        return True
    except Exception:
        print("GLUT sem suporte a retorno no close; fechar pode encerrar o kernel.")
        return False

def draw_square() -> None:
    glBegin(GL_QUADS)
    glColor3f(0.2, 0.75, 1.0)
    glVertex2f(-0.5, -0.5); glVertex2f(0.5, -0.5)
    glVertex2f(0.5, 0.5); glVertex2f(-0.5, 0.5)
    glEnd()

def display() -> None:
    glClear(GL_COLOR_BUFFER_BIT)
    glMatrixMode(GL_MODELVIEW)
    glLoadIdentity()
    draw_square()
    glFlush()

def reshape(width: int, height: int) -> None:
    if height == 0:
        height = 1
    aspect = width / height

    glViewport(0, 0, width, height)
    glMatrixMode(GL_PROJECTION)
    glLoadIdentity()

    if aspect >= 1.0:
        gluOrtho2D(-aspect, aspect, -1.0, 1.0)
    else:
        inv = 1.0 / aspect
        gluOrtho2D(-1.0, 1.0, -inv, inv)
```

```python
glutInit()
close_safe = configure_close_behavior()
glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB)
glutInitWindowSize(900, 600)
glutCreateWindow(b"Aula 05 - Reshape e Viewport")
glClearColor(0.05, 0.05, 0.07, 1.0)
glutDisplayFunc(display)
glutReshapeFunc(reshape)

try:
    glutMainLoop()
except SystemExit:
    print("GLUT chamou SystemExit ao fechar a janela.")

if close_safe:
    print("Janela fechada e kernel deve continuar ativo.")
else:
    print("Se o kernel cair, rode esta aula via script .py.")
```
