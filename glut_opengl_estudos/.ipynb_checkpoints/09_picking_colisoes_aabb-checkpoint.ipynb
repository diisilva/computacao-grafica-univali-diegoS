# Aula 09 — Picking por Mouse + Colisão AABB

Exemplo profissional de lógica de interação e física simples em tempo real.

## Objetivos de engenharia
- Representar entidades com estrutura de dados clara (`dataclass`).
- Realizar seleção por clique no espaço NDC.
- Detectar colisões AABB com feedback visual.

## Conceitos-chave
- Conversão pixel → NDC para picking.
- AABB (`Axis-Aligned Bounding Box`) para colisão eficiente.
- Loop de atualização separado da renderização.

## Controles
- Clique esquerdo: selecionar caixa.
- `R`: reset.
- `ESC`: sair.

```python
from dataclasses import dataclass
from random import Random
from OpenGL.GL import *
from OpenGL.GLUT import *

@dataclass
class BoxEntity:
    """Caixa com posição, velocidade e seleção."""
    x: float
    y: float
    half_size: float
    vx: float
    vy: float
    color: tuple[float, float, float]
    selected: bool = False

    def min_x(self) -> float: return self.x - self.half_size
    def max_x(self) -> float: return self.x + self.half_size
    def min_y(self) -> float: return self.y - self.half_size
    def max_y(self) -> float: return self.y + self.half_size

    def contains(self, px: float, py: float) -> bool:
        return self.min_x() <= px <= self.max_x() and self.min_y() <= py <= self.max_y()

class Demo:
    def __init__(self) -> None:
        self.rng = Random(42)
        self.entities: list[BoxEntity] = []
        self.colliding_pairs: set[tuple[int, int]] = set()
        self.reset()

    def reset(self) -> None:
        self.entities.clear()
        palette = [(1.0,0.4,0.4),(0.4,1.0,0.5),(0.4,0.7,1.0),(1.0,0.9,0.4),(0.9,0.5,1.0)]
        for i in range(6):
            self.entities.append(BoxEntity(self.rng.uniform(-0.8,0.8), self.rng.uniform(-0.6,0.6), self.rng.uniform(0.07,0.11), self.rng.uniform(-0.006,0.006), self.rng.uniform(-0.006,0.006), palette[i % len(palette)]))

    def overlap(self, a: BoxEntity, b: BoxEntity) -> bool:
        return not (a.max_x() < b.min_x() or a.min_x() > b.max_x() or a.max_y() < b.min_y() or a.min_y() > b.max_y())

    def update(self) -> None:
        for e in self.entities:
            e.x += e.vx
            e.y += e.vy
            if e.min_x() < -1.0 or e.max_x() > 1.0: e.vx *= -1
            if e.min_y() < -1.0 or e.max_y() > 1.0: e.vy *= -1
            e.x = max(-1 + e.half_size, min(1 - e.half_size, e.x))
            e.y = max(-1 + e.half_size, min(1 - e.half_size, e.y))

        self.colliding_pairs.clear()
        for i in range(len(self.entities)):
            for j in range(i + 1, len(self.entities)):
                if self.overlap(self.entities[i], self.entities[j]):
                    self.colliding_pairs.add((i, j))

    def pick_at(self, px: float, py: float) -> None:
        for e in self.entities: e.selected = False
        for e in reversed(self.entities):
            if e.contains(px, py):
                e.selected = True
                break

    def render(self) -> None:
        glClear(GL_COLOR_BUFFER_BIT)
        glMatrixMode(GL_MODELVIEW)
        glLoadIdentity()
        coll = {i for pair in self.colliding_pairs for i in pair}
        for i, e in enumerate(self.entities):
            glColor3f(1.0, 0.2, 0.2) if i in coll else glColor3f(*e.color)
            glBegin(GL_QUADS)
            glVertex2f(e.min_x(), e.min_y())
            glVertex2f(e.max_x(), e.min_y())
            glVertex2f(e.max_x(), e.max_y())
            glVertex2f(e.min_x(), e.max_y())
            glEnd()
            if e.selected:
                glColor3f(1.0, 1.0, 1.0)
                glBegin(GL_LINE_LOOP)
                glVertex2f(e.min_x()-0.01, e.min_y()-0.01)
                glVertex2f(e.max_x()+0.01, e.min_y()-0.01)
                glVertex2f(e.max_x()+0.01, e.max_y()+0.01)
                glVertex2f(e.min_x()-0.01, e.max_y()+0.01)
                glEnd()
        glutSwapBuffers()

demo = Demo()

def configure_close_behavior() -> bool:
    try:
        glutSetOption(GLUT_ACTION_ON_WINDOW_CLOSE, GLUT_ACTION_GLUTMAINLOOP_RETURNS)
        return True
    except Exception:
        return False

def display() -> None: demo.render()

def keyboard(key: bytes, _x: int, _y: int) -> None:
    if key in (b'r', b'R'): demo.reset()
    elif key == b'\x1b': raise SystemExit('ESC')

def mouse(button: int, state: int, x: int, y: int) -> None:
    if button != GLUT_LEFT_BUTTON or state != GLUT_DOWN: return
    w = glutGet(GLUT_WINDOW_WIDTH)
    h = glutGet(GLUT_WINDOW_HEIGHT)
    px = 2.0 * (x / w) - 1.0
    py = 1.0 - 2.0 * (y / h)
    demo.pick_at(px, py)
    glutPostRedisplay()

def tick(_v: int) -> None:
    demo.update()
    glutPostRedisplay()
    glutTimerFunc(16, tick, 0)
```

```python
glutInit()
close_safe = configure_close_behavior()
glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB)
glutInitWindowSize(1200, 850)
glutCreateWindow(b"Aula 09 - Picking e Colisoes AABB")
glClearColor(0.04, 0.04, 0.07, 1.0)
glutDisplayFunc(display)
glutKeyboardFunc(keyboard)
glutMouseFunc(mouse)
glutTimerFunc(0, tick, 0)

try:
    glutMainLoop()
except SystemExit:
    print("GLUT chamou SystemExit ao fechar a janela.")

print("Kernel preservado:", close_safe)
```
