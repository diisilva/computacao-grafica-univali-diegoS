# Aula 08 — Scene Graph Hierárquico 2D (Nível Profissional)

Este notebook aplica um design de engenharia para cenas OpenGL com **árvore de nós** e **transformações locais**.

## Por que isso é senior-level
- Separação clara entre estado da cena, atualização e render.
- Reuso de comportamento via nós recursivos.
- Base sólida para evoluir para motores 2D/3D mais complexos.

## Conceitos-chave
- `Transform2D`: estado local de transformação.
- `SceneNode`: composição hierárquica de objetos.
- Render recursivo com pilha de matrizes.

## Execução
1. Rode a célula 2 para carregar arquitetura e callbacks.
2. Rode a célula 3 para abrir a janela com proteção de kernel.

```python
from dataclasses import dataclass, field
from math import cos, sin, tau
from typing import Callable

from OpenGL.GL import *
from OpenGL.GLUT import *

@dataclass
class Transform2D:
    """Transformação local de um nó na cena."""
    tx: float = 0.0
    ty: float = 0.0
    rotation_deg: float = 0.0
    sx: float = 1.0
    sy: float = 1.0

@dataclass
class SceneNode:
    """Nó de scene graph com draw opcional e filhos."""
    name: str
    transform: Transform2D = field(default_factory=Transform2D)
    draw_fn: Callable[[], None] | None = None
    children: list['SceneNode'] = field(default_factory=list)

    def render(self) -> None:
        glPushMatrix()
        glTranslatef(self.transform.tx, self.transform.ty, 0.0)
        glRotatef(self.transform.rotation_deg, 0.0, 0.0, 1.0)
        glScalef(self.transform.sx, self.transform.sy, 1.0)

        if self.draw_fn is not None:
            self.draw_fn()

        for child in self.children:
            child.render()
        glPopMatrix()

class SceneGraphDemo:
    """Demo de arquitetura em árvore com órbita dupla."""
    def __init__(self) -> None:
        self.time = 0.0
        self.speed = 1.0
        self.paused = False

        self.root = SceneNode('root')
        self.pivot_outer = SceneNode('pivot_outer')
        self.pivot_inner = SceneNode('pivot_inner')

        self.core = SceneNode('core', draw_fn=self._draw_core)
        self.outer = SceneNode('outer', draw_fn=self._draw_outer)
        self.inner = SceneNode('inner', draw_fn=self._draw_inner)

        self.root.children.extend([self.core, self.pivot_outer])
        self.pivot_outer.children.extend([self.outer, self.pivot_inner])
        self.pivot_inner.children.append(self.inner)

    def _circle(self, r: float, color: tuple[float, float, float], seg: int = 40) -> None:
        glColor3f(*color)
        glBegin(GL_POLYGON)
        for i in range(seg):
            a = tau * i / seg
            glVertex2f(r * cos(a), r * sin(a))
        glEnd()

    def _orbit(self, r: float) -> None:
        glColor3f(0.2, 0.2, 0.25)
        glBegin(GL_LINE_LOOP)
        for i in range(96):
            a = tau * i / 96
            glVertex2f(r * cos(a), r * sin(a))
        glEnd()

    def _draw_core(self) -> None:
        self._circle(0.10, (1.0, 0.85, 0.2), 60)

    def _draw_outer(self) -> None:
        self._circle(0.035, (0.25, 0.6, 1.0), 36)

    def _draw_inner(self) -> None:
        self._circle(0.015, (0.85, 0.85, 0.95), 24)

    def update(self) -> None:
        if not self.paused:
            self.time += 0.012 * self.speed
        self.pivot_outer.transform.rotation_deg = self.time * 80.0
        self.outer.transform.tx = 0.52
        self.pivot_inner.transform.rotation_deg = self.time * 240.0
        self.inner.transform.tx = 0.14

    def render(self) -> None:
        glClear(GL_COLOR_BUFFER_BIT)
        glMatrixMode(GL_MODELVIEW)
        glLoadIdentity()

        self._orbit(0.52)
        glPushMatrix()
        glRotatef(self.pivot_outer.transform.rotation_deg, 0.0, 0.0, 1.0)
        glTranslatef(0.52, 0.0, 0.0)
        self._orbit(0.14)
        glPopMatrix()

        self.root.render()
        glutSwapBuffers()

demo = SceneGraphDemo()

def configure_close_behavior() -> bool:
    try:
        glutSetOption(GLUT_ACTION_ON_WINDOW_CLOSE, GLUT_ACTION_GLUTMAINLOOP_RETURNS)
        return True
    except Exception:
        return False

def display() -> None:
    demo.render()

def keyboard(key: bytes, _x: int, _y: int) -> None:
    if key == b' ':
        demo.paused = not demo.paused
    elif key == b'+':
        demo.speed = min(demo.speed + 0.2, 6.0)
    elif key == b'-':
        demo.speed = max(demo.speed - 0.2, 0.2)
    elif key == b'\x1b':
        raise SystemExit('ESC')

def tick(_v: int) -> None:
    demo.update()
    glutPostRedisplay()
    glutTimerFunc(16, tick, 0)
```

```python
glutInit()
close_safe = configure_close_behavior()
glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB)
glutInitWindowSize(1100, 800)
glutCreateWindow(b"Aula 08 - Scene Graph 2D")
glClearColor(0.03, 0.03, 0.06, 1.0)
glutDisplayFunc(display)
glutKeyboardFunc(keyboard)
glutTimerFunc(0, tick, 0)

try:
    glutMainLoop()
except SystemExit:
    print("GLUT chamou SystemExit ao fechar a janela.")

print("Kernel preservado:", close_safe)
```
