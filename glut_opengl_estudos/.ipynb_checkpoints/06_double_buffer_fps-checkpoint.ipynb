# Aula 06 — Double Buffer e Medição de FPS

> Objetivo: entender por que `GLUT_DOUBLE` é mais suave e como medir FPS.

## Conceitos importantes
- **Double buffering**: desenha no back buffer e depois troca (`glutSwapBuffers`).
- **Idle callback** (`glutIdleFunc`): atualiza continuamente a animação.
- **FPS médio**: razão entre número de frames e tempo decorrido.

## Roteiro de execução
1. Execute a célula 1.
2. Execute a célula 2 e observe FPS no console.

```python
from time import perf_counter
from OpenGL.GL import *
from OpenGL.GLUT import *

angle = 0.0
frame_count = 0
start_time = perf_counter()

def configure_close_behavior() -> bool:
    """Tenta evitar encerramento do kernel ao fechar a janela GLUT."""
    try:
        glutSetOption(GLUT_ACTION_ON_WINDOW_CLOSE, GLUT_ACTION_GLUTMAINLOOP_RETURNS)
        print("GLUT configurado para retornar ao fechar a janela.")
        return True
    except Exception:
        print("GLUT sem suporte a retorno no close; fechar pode encerrar o kernel.")
        return False

def draw_rotating_quad() -> None:
    glBegin(GL_QUADS)
    glColor3f(1.0, 0.6, 0.2); glVertex2f(-0.3, -0.3); glVertex2f(0.3, -0.3)
    glColor3f(0.2, 0.8, 1.0); glVertex2f(0.3, 0.3); glVertex2f(-0.3, 0.3)
    glEnd()

def display() -> None:
    global frame_count, start_time
    glClear(GL_COLOR_BUFFER_BIT)
    glLoadIdentity()
    glRotatef(angle, 0.0, 0.0, 1.0)
    draw_rotating_quad()
    glutSwapBuffers()

    frame_count += 1
    elapsed = perf_counter() - start_time
    if elapsed >= 1.0:
        print(f'FPS medio: {frame_count / elapsed:.1f}')
        frame_count = 0
        start_time = perf_counter()

def idle() -> None:
    global angle
    angle = (angle + 0.35) % 360.0
    glutPostRedisplay()
```

```python
glutInit()
close_safe = configure_close_behavior()
glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB)
glutInitWindowSize(900, 600)
glutCreateWindow(b"Aula 06 - Double Buffer e FPS")
glClearColor(0.04, 0.04, 0.06, 1.0)
glutDisplayFunc(display)
glutIdleFunc(idle)

try:
    glutMainLoop()
except SystemExit:
    print("GLUT chamou SystemExit ao fechar a janela.")

if close_safe:
    print("Janela fechada e kernel deve continuar ativo.")
else:
    print("Se o kernel cair, rode esta aula via script .py.")
```
