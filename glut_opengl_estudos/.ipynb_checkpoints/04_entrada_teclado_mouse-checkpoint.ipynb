# Aula 04 — Entrada com Teclado e Mouse

> Objetivo: controlar um objeto 2D em tempo real com eventos do GLUT.

## Conceitos importantes
- Callback de teclado (`glutKeyboardFunc`).
- Callback de mouse (`glutMouseFunc`).
- Conversão de pixel para NDC para posicionar o objeto.

## Controles
- `W`, `A`, `S`, `D`: movimentar quadrado.
- `R`, `G`, `B`: mudar cor.
- Clique esquerdo: mover para posição do cursor.
- `ESC`: encerrar.

## Roteiro de execução
1. Execute a célula 1.
2. Execute a célula 2.

```python
from OpenGL.GL import *
from OpenGL.GLUT import *

position_x = 0.0
position_y = 0.0
current_color = (0.9, 0.4, 0.2)

def configure_close_behavior() -> bool:
    """Tenta evitar encerramento do kernel ao fechar a janela GLUT."""
    try:
        glutSetOption(GLUT_ACTION_ON_WINDOW_CLOSE, GLUT_ACTION_GLUTMAINLOOP_RETURNS)
        print("GLUT configurado para retornar ao fechar a janela.")
        return True
    except Exception:
        print("GLUT sem suporte a retorno no close; fechar pode encerrar o kernel.")
        return False

def draw_square() -> None:
    glBegin(GL_QUADS)
    glVertex2f(-0.15, -0.15); glVertex2f(0.15, -0.15)
    glVertex2f(0.15, 0.15); glVertex2f(-0.15, 0.15)
    glEnd()

def display() -> None:
    glClear(GL_COLOR_BUFFER_BIT)
    glLoadIdentity()
    glTranslatef(position_x, position_y, 0.0)
    glColor3f(*current_color)
    draw_square()
    glFlush()

def keyboard(key: bytes, _: int, __: int) -> None:
    global position_x, position_y, current_color
    step = 0.05
    if key == b'w': position_y += step
    elif key == b's': position_y -= step
    elif key == b'a': position_x -= step
    elif key == b'd': position_x += step
    elif key == b'r': current_color = (1.0, 0.2, 0.2)
    elif key == b'g': current_color = (0.2, 1.0, 0.2)
    elif key == b'b': current_color = (0.2, 0.4, 1.0)
    elif key == b'\x1b': raise SystemExit('ESC pressionado')
    glutPostRedisplay()

def mouse(button: int, state: int, x: int, y: int) -> None:
    global position_x, position_y
    if button != GLUT_LEFT_BUTTON or state != GLUT_DOWN:
        return
    width = glutGet(GLUT_WINDOW_WIDTH)
    height = glutGet(GLUT_WINDOW_HEIGHT)
    position_x = 2.0 * (x / width) - 1.0
    position_y = 1.0 - 2.0 * (y / height)
    glutPostRedisplay()
```

```python
glutInit()
close_safe = configure_close_behavior()
glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB)
glutInitWindowSize(900, 600)
glutCreateWindow(b"Aula 04 - Teclado e Mouse")
glClearColor(0.07, 0.07, 0.1, 1.0)
glutDisplayFunc(display)
glutKeyboardFunc(keyboard)
glutMouseFunc(mouse)

try:
    glutMainLoop()
except SystemExit:
    print("GLUT chamou SystemExit ao fechar a janela.")

if close_safe:
    print("Janela fechada e kernel deve continuar ativo.")
else:
    print("Se o kernel cair, rode esta aula via script .py.")
```
