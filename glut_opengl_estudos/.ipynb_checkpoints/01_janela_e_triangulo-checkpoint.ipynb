# Aula 01 — Primeira Janela com GLUT + OpenGL

Este é o ponto de partida da trilha. Aqui você abre uma janela OpenGL e desenha seu primeiro triângulo.

## Conceitos-chave
- **GLUT**: cria janela, registra callbacks e gerencia loop de eventos.
- **Callback `display`**: função chamada para desenhar a cena.
- **`GL_TRIANGLES`**: desenha um triângulo a partir de 3 vértices.
- **Interpolação de cor**: cada vértice recebe uma cor e o interior mistura essas cores.

## Roteiro de execução
1. Execute a célula 2 para importar, definir `display` e configurar o GLUT.
2. A janela abrirá e mostrará o triângulo colorido.

## Observação
Em alguns ambientes, fechar a janela GLUT pode encerrar o kernel do notebook. Se ocorrer, reinicie o kernel e rode novamente.

```python
from OpenGL.GL import *
from OpenGL.GLUT import *

def configure_close_behavior() -> bool:
    """Tenta evitar encerramento do kernel ao fechar a janela GLUT."""
    try:
        glutSetOption(GLUT_ACTION_ON_WINDOW_CLOSE, GLUT_ACTION_GLUTMAINLOOP_RETURNS)
        print("GLUT configurado para retornar ao fechar a janela.")
        return True
    except Exception:
        print("GLUT sem suporte a retorno no close; fechar pode encerrar o kernel.")
        return False

def display() -> None:
    """Desenha um triângulo RGB."""
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)
    glBegin(GL_TRIANGLES)
    glColor3f(1.0, 0.0, 0.0); glVertex2f(-0.7, -0.6)
    glColor3f(0.0, 1.0, 0.0); glVertex2f(0.7, -0.6)
    glColor3f(0.0, 0.2, 1.0); glVertex2f(0.0, 0.7)
    glEnd()
    glFlush()

glutInit()
close_safe = configure_close_behavior()
glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB | GLUT_DEPTH)
glutInitWindowSize(800, 600)
glutCreateWindow(b"Aula 01 - Triangulo")
glClearColor(0.05, 0.05, 0.08, 1.0)
glutDisplayFunc(display)

try:
    glutMainLoop()
except SystemExit:
    print("GLUT chamou SystemExit ao fechar a janela.")

if close_safe:
    print("Janela fechada e kernel deve continuar ativo.")
else:
    print("Se o kernel cair, rode esta aula via script .py.")
```
