# Aula 07 — Mini Projeto: Sistema Solar 2D Completo

> Objetivo: integrar animação, hierarquia de transformações, texto em tela e interação.

## O que foi adicionado
- Sol + 8 planetas: Mercúrio, Vênus, Terra, Marte, Júpiter, Saturno, Urano e Netuno.
- Campo de estrelas no fundo.
- Órbitas visuais e legenda com nome dos planetas.
- Controles para velocidade, pausa e mostrar/ocultar rótulos.

## Controles
- `+`: aumenta velocidade orbital.
- `-`: diminui velocidade orbital.
- `Espaço`: pausa/retoma.
- `L`: liga/desliga rótulos dos planetas.
- `ESC`: encerra.

## Roteiro
1. Execute a célula 2 para carregar funções e estado da simulação.
2. Execute a célula 3 para iniciar a animação GLUT.

```python
from math import cos, sin, tau
from OpenGL.GL import *
from OpenGL.GLUT import *

sim_time = 0.0
speed = 1.0
paused = False
show_labels = True

PLANETS = [
    {"name": "Mercurio", "orbit": 0.14, "radius": 0.010, "speed": 4.8, "color": (0.70, 0.70, 0.70)},
    {"name": "Venus",    "orbit": 0.20, "radius": 0.016, "speed": 3.5, "color": (0.90, 0.75, 0.45)},
    {"name": "Terra",    "orbit": 0.27, "radius": 0.017, "speed": 3.0, "color": (0.20, 0.55, 1.00)},
    {"name": "Marte",    "orbit": 0.34, "radius": 0.014, "speed": 2.4, "color": (0.90, 0.35, 0.25)},
    {"name": "Jupiter",  "orbit": 0.48, "radius": 0.040, "speed": 1.3, "color": (0.85, 0.65, 0.45)},
    {"name": "Saturno",  "orbit": 0.62, "radius": 0.033, "speed": 1.0, "color": (0.90, 0.82, 0.55)},
    {"name": "Urano",    "orbit": 0.76, "radius": 0.024, "speed": 0.7, "color": (0.55, 0.85, 0.90)},
    {"name": "Netuno",   "orbit": 0.88, "radius": 0.023, "speed": 0.55, "color": (0.30, 0.45, 0.95)},
]

STARS = [
    (-0.95, 0.95), (-0.82, 0.75), (-0.60, 0.90), (-0.40, 0.70), (-0.20, 0.96),
    (0.05, 0.88), (0.28, 0.74), (0.52, 0.93), (0.74, 0.79), (0.92, 0.91),
    (-0.90, 0.45), (-0.65, 0.52), (-0.42, 0.40), (-0.15, 0.55), (0.16, 0.46),
    (0.43, 0.51), (0.70, 0.44), (0.90, 0.36), (-0.85, 0.10), (-0.58, 0.20),
    (-0.30, 0.12), (-0.05, 0.18), (0.20, 0.08), (0.48, 0.16), (0.77, 0.12),
    (-0.94, -0.24), (-0.66, -0.15), (-0.39, -0.29), (-0.12, -0.18), (0.14, -0.25),
    (0.38, -0.12), (0.64, -0.22), (0.90, -0.30), (-0.78, -0.60), (-0.50, -0.52),
    (-0.24, -0.72), (0.02, -0.64), (0.28, -0.77), (0.56, -0.58), (0.84, -0.70),
]

def configure_close_behavior() -> bool:
    """Tenta evitar encerramento do kernel ao fechar a janela GLUT."""
    try:
        glutSetOption(GLUT_ACTION_ON_WINDOW_CLOSE, GLUT_ACTION_GLUTMAINLOOP_RETURNS)
        print("GLUT configurado para retornar ao fechar a janela.")
        return True
    except Exception:
        print("GLUT sem suporte a retorno no close; fechar pode encerrar o kernel.")
        return False

def draw_text(x: float, y: float, text: str) -> None:
    """Desenha texto 2D usando fonte bitmap do GLUT."""
    glRasterPos2f(x, y)
    for char in text:
        glutBitmapCharacter(GLUT_BITMAP_HELVETICA_12, ord(char))

def draw_circle(radius: float, color: tuple[float, float, float], segments: int = 40) -> None:
    """Desenha um círculo preenchido aproximado por polígono."""
    glColor3f(*color)
    glBegin(GL_POLYGON)
    for i in range(segments):
        t = tau * i / segments
        glVertex2f(radius * cos(t), radius * sin(t))
    glEnd()

def draw_orbit(radius: float) -> None:
    """Desenha a órbita circular de um planeta."""
    glColor3f(0.18, 0.18, 0.25)
    glBegin(GL_LINE_LOOP)
    for i in range(120):
        t = tau * i / 120
        glVertex2f(radius * cos(t), radius * sin(t))
    glEnd()

def draw_stars() -> None:
    """Desenha estrelas fixas no fundo da cena."""
    glPointSize(2.0)
    glColor3f(0.95, 0.95, 1.0)
    glBegin(GL_POINTS)
    for sx, sy in STARS:
        glVertex2f(sx, sy)
    glEnd()

def display() -> None:
    """Renderiza sistema solar 2D com legenda opcional."""
    glClear(GL_COLOR_BUFFER_BIT)

    draw_stars()

    for planet in PLANETS:
        draw_orbit(planet["orbit"])

    draw_circle(0.08, (1.0, 0.85, 0.15), segments=60)

    for planet in PLANETS:
        angle = sim_time * planet["speed"]
        x = planet["orbit"] * cos(angle)
        y = planet["orbit"] * sin(angle)

        glPushMatrix()
        glTranslatef(x, y, 0.0)
        draw_circle(planet["radius"], planet["color"], segments=28)
        glPopMatrix()

        if show_labels:
            glColor3f(0.85, 0.9, 1.0)
            draw_text(x + planet["radius"] + 0.01, y + planet["radius"] + 0.01, planet["name"])

    glColor3f(1.0, 1.0, 0.6)
    draw_text(-0.96, -0.95, "Controles: +  -  Espaco  L  ESC")

    glutSwapBuffers()

def update(_: int) -> None:
    """Atualiza o tempo da simulação e agenda o próximo frame."""
    global sim_time
    if not paused:
        sim_time += 0.01 * speed
    glutPostRedisplay()
    glutTimerFunc(16, update, 0)

def keyboard(key: bytes, _: int, __: int) -> None:
    """Controla velocidade, pausa, legenda e saída."""
    global speed, paused, show_labels
    if key == b'+':
        speed = min(speed + 0.2, 8.0)
    elif key == b'-':
        speed = max(speed - 0.2, 0.2)
    elif key == b' ':
        paused = not paused
    elif key in (b'l', b'L'):
        show_labels = not show_labels
    elif key == b'\x1b':
        raise SystemExit('ESC')

    print(f"speed={speed:.1f} | paused={paused} | labels={show_labels}")
```

```python
glutInit()
glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB)
glutInitWindowSize(950, 700)
glutCreateWindow(b"Aula 07 - Sistema Solar 2D")
glClearColor(0.01, 0.01, 0.05, 1.0)
glutDisplayFunc(display)
glutKeyboardFunc(keyboard)
glutTimerFunc(0, update, 0)
glutMainLoop()


try:
    glutMainLoop()
except SystemExit:
    print("GLUT chamou SystemExit ao fechar a janela.")

if close_safe:
    print("Janela fechada e kernel deve continuar ativo.")
else:
    print("Se o kernel cair, rode esta aula via script .py.")
```
