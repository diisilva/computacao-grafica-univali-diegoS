```python
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
```

```python
def display():
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)
    glBegin(GL_TRIANGLES)
    glColor3f(1, 0, 0); glVertex2f(-0.5, -0.5)
    glColor3f(0, 1, 0); glVertex2f(0.5, -0.5)
    glColor3f(0, 0, 1); glVertex2f(0.0, 0.5)
    glEnd()
    glFlush()

def configure_glut_close_behavior():
    try:
        glutSetOption(GLUT_ACTION_ON_WINDOW_CLOSE, GLUT_ACTION_GLUTMAINLOOP_RETURNS)
        print("GLUT configurado para retornar ao fechar a janela.")
        return True
    except Exception:
        print("GLUT sem suporte a retorno no close; fechar a janela pode encerrar o kernel.")
        return False
```

```python
glutInit()
close_safe = configure_glut_close_behavior()

glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB)
glutInitWindowSize(400, 400)
glutCreateWindow(b"Teste GLUT / PyOpenGL")
glutDisplayFunc(display)
glClearColor(0.0, 0.0, 0.0, 1.0)

try:
    glutMainLoop()
except SystemExit:
    print("GLUT chamou SystemExit ao fechar a janela.")

if close_safe:
    print("Janela fechada e kernel continua ativo.")
else:
    print("Se o kernel cair ao fechar, rode OpenGL em processo separado.")
```
